```python
from transformers import pipeline

clf = pipeline(
    "text-classification",
    model="LiYuan/amazon-review-sentiment-analysis",
    device=0,
    truncation=True,
    max_length=256,
)

test_sentences = [
    "This product is fantastic and works perfectly",
    "Arrived broken, total waste of money",
    "It is okay, nothing special but does the job",
    "Absolutely love it! Will buy again.",
    "Stopped working after two days. Very disappointed.",
]

LABEL2STAR = {
    "1 star": 1, "2 stars": 2, "3 stars": 3, "4 stars": 4, "5 stars": 5
}
STAR2LABEL = {
    1: "Negative", 2: "Moderately Negative", 3: "Neutral", 4: "Satisfactory", 5: "Positive"
}

for sentence in test_sentences:
    result = clf(sentence)[0]
    star = LABEL2STAR[result["label"]]        # "5 stars" → 5
    stars = "★" * star + "☆" * (5 - star)
    print(f"Input      : {sentence}")
    print(f"Rating     : {star}/5 {stars}")
    print(f"Label      : {STAR2LABEL[star]}")
    print(f"Confidence : {round(result['score'], 4)}\n")
```
